```python
wellsfargoleaderboard21 = pd.read_html("https://www.espn.com/golf/leaderboard?tournamentId=401243420")[0]

wellsfargoleaderboard21
```

```python
# Need to manually find the name of the first person cut on the leaderboard so we can remove all the people who were cut from the board. In this case it was harry higgs

wellsfargoleaderboard21 = wellsfargoleaderboard21[0:77]
wellsfargoleaderboard21 = wellsfargoleaderboard21[['PLAYER', 'TOT']]
wellsfargoleaderboard21.columns = ['PLAYER NAME', 'TOT']
# This final leaderboard is the dependant variable for our regression. In our regression we want the coefficients to be as negative as possible since we are trying to maximize the correlation between positive strokes gained but a lower score.
wellsfargoleaderboard21
```

```python
dfforregression = wellsfargoleaderboard21.merge(dfmaster21, on="PLAYER NAME", how="left")
df = dfforregression 
df = df.drop([54, 57, 60, 66])
df
```

```python
# Define X and Y
x = df[['AVERAGE_putting_scaled', 'AVERAGE_OTT_scaled', 'AVERAGE_ARTG_scaled', 'AVERAGE_ATG_scaled']]
y = df['TOT']
```

```python
regr = linear_model.LinearRegression()
regr.fit(x,y)
```

```python
# Getting Each Coefficient
co = regr.coef_
Putting = co[0]
OTT = co[1]
ARTG = co[2]
ATG = co[3]

print(f'The Coefficient for putting is {Putting}')
print(f'The Coefficient for OTT is {OTT}')
print(f'The Coefficient for ARTG is {ARTG}')
print(f'The Coefficient for ATG is {ATG}')

sumofcoefficients = float(Putting) + float(OTT) + float(ARTG) + float(ATG)
print('The sum of coefficients is', sumofcoefficients)
```

```python
# Find each contribution by putting the coefficient over the total
puttingcont = float(Putting)/sumofcoefficients
OTTcont = float(OTT)/sumofcoefficients
ARTGcont = float(ARTG)/sumofcoefficients
ATGcont = float(ATG)/sumofcoefficients

print(f'The putting contribution is {puttingcont}')
print(f'The OTT contribution is {OTTcont}')
print(f'The ARTG contribution is {ARTGcont}')
print(f'The ATG contribution is {ATGcont}')
```
