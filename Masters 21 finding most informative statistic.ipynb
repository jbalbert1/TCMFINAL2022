### Masters Backtest finding most informative course statistic

```python
import pandas as pd
from sklearn import linear_model
```

```python
%store -r dfmaster21
```

```python
dfmaster21
```

```python
mastersleaderboard21 = pd.read_html("https://www.espn.com/golf/leaderboard/_/tournamentId/401243010")[0]
pd.set_option('display.max_rows', None)
mastersleaderboard21
```

```python
mastersleaderboard21 = mastersleaderboard21[0:54]
mastersleaderboard21 = mastersleaderboard21[['PLAYER', 'TOT']]
mastersleaderboard21.columns = ['PLAYER NAME', 'TOT']
# This final leaderboard is the dependant variable for our regression. In our regression we want the coefficients to be as negative as possible since we are trying to maximize the correlation between positive strokes gained but a lower score.
mastersleaderboard21
```

```python
dfforregression = mastersleaderboard21.merge(dfmaster21, on="PLAYER NAME", how="left")
df = dfforregression
df = df.drop([15, 38, 39, 44, 49, 51])
df
```

```python
# Define X and Y
x = df[['AVERAGE_putting_scaled', 'AVERAGE_OTT_scaled', 'AVERAGE_ARTG_scaled', 'AVERAGE_ATG_scaled']]
y = df['TOT']
```

```python
regr = linear_model.LinearRegression()
regr.fit(x,y)
```

```python
# Getting Each Coefficient
co = regr.coef_
Puttingmasters = co[0]
OTTmasters = co[1]
ARTGmasters = co[2]
ATGmasters = co[3]

print(f'The Coefficient for putting is {Puttingmasters}')
print(f'The Coefficient for OTT is {OTTmasters}')
print(f'The Coefficient for ARTG is {ARTGmasters}')
print(f'The Coefficient for ATG is {ATGmasters}')

sumofcoefficients = float(Puttingmasters) + float(OTTmasters) + float(ARTGmasters) + float(ATGmasters)
print('The sum of coefficients is', sumofcoefficients)
```

```python
## Since some of the coefficients are positive here, 
## which we don't want, we are going to subtract the absolute value of the most positive statistic
## from each of the values so that we have no positives
## Then we need to reprint each coefficient and the sum

adjustmentnumber = 2.4276054379815197

Puttingmasters = Puttingmasters - adjustmentnumber
OTTmasters = OTTmasters - adjustmentnumber
ARTGmasters = ARTGmasters - adjustmentnumber
ATGmasters = ATGmasters - adjustmentnumber

print(f'The adjusted Coefficient for putting is {Puttingmasters}')
print(f'The adjusted Coefficient for OTT is {OTTmasters}')
print(f'The adjusted Coefficient for ARTG is {ARTGmasters}')
print(f'The adjusted Coefficient for ATG is {ATGmasters}')

sumofcoefficients = float(Puttingmasters) + float(OTTmasters) + float(ARTGmasters) + float(ATGmasters)
print('The adjusted sum of coefficients is', sumofcoefficients)
```

```python
# There we go
# Now find each contribution by putting the coefficient over the total
puttingcontmasters = float(Puttingmasters)/sumofcoefficients
OTTcontmasters = float(OTTmasters)/sumofcoefficients
ARTGcontmasters = float(ARTGmasters)/sumofcoefficients
ATGcontmasters = float(ATGmasters)/sumofcoefficients

print(f'The putting contribution is {puttingcontmasters}')
print(f'The OTT contribution is {OTTcontmasters}')
print(f'The ARTG contribution is {ARTGcontmasters}')
print(f'The ATG contribution is {ATGcontmasters}')
```
